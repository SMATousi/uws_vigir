```python
import os
import pickle
import pandas as pd

# smalltrue
df = pd.DataFrame()
for fname in os.listdir('.'):
    if fname.startswith('smalltrue'):
        with open(fname, 'rb') as f:
            result = pickle.load(f)
        sp = fname.split('_')
        n_train = sp[1]
        seed = sp[2].split('.')[0]
        test_tau = min(result['test_tau'])[0]
        df = df.append({
            'condition': 'fully_supervised',
            'n_train': int(n_train),
            'seed': seed,
            'test_tau': test_tau
        }, ignore_index=True)
```
